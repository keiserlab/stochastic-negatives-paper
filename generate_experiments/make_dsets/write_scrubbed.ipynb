```python
import numpy as np
import pandas as pd
import os
```

```python
def get_env_var(handle):
    tmp = os.getenv(handle)
    if not tmp:
        raise LookupError("Environment variable: {} not set.".format(handle))
    return tmp.strip("'")
```

```python
base = get_env_var("DATA_SAVE_BASE")
dm_base = "/srv/home/ecaceres/chembl20_minibatching/datasets/raw_data/from_mysql_queries/drug_mat"
dm_mids = "{}/drug_mat_chembl20_mwcut800_unique_mids.txt".format(dm_base)
dm_tids = "{}/drug_mat_chembl20_mwcut800_unique_tids.txt".format(dm_base)
dm_dat = "{}/drug_matrix_chembl20_mwcutoff800.csv.gz".format(dm_base)
```

```python
df = pd.read_csv(dm_dat, sep="\t")
```

```python
pairs = df[["ChEMBL_Target_ID", "ChEMBL_Molecule_ID",]]
```

```python
tid_mid_pairs = {(p["ChEMBL_Target_ID"],p["ChEMBL_Molecule_ID"]) for idx, p in pairs.iterrows()}    
```

```python
chembl_dat = "/srv/home/ecaceres/chembl20_minibatching/datasets/raw_data/from_mysql_queries/full_chembl20_cutoff800_sorted_by_mid_tid.csv.gz"
```

```python
c_df = pd.read_csv(chembl_dat, compression="gzip", sep="\t")
```

```python
c_pairs = c_df[["ChEMBL_Target_ID", "ChEMBL_Molecule_ID",]]
```

```python
row_blacklist = [idx for idx, p in c_pairs.iterrows() if (p["ChEMBL_Target_ID"],p["ChEMBL_Molecule_ID"]) in tid_mid_pairs]
```

```python
new_df = c_df.drop(row_blacklist, axis=0)
```

```python
save_base = "{}".format(base)
save_name = "{}/raw_data/full_chembl20_cutoff800_dm_scrubbed.csv.gz".format(save_base)
```

```python
new_df.to_csv(save_name, sep="\t", compression="gzip", header=True, index=False)
```
