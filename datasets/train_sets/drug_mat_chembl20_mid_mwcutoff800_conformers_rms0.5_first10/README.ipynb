This directory contains our drug matrix validation set to be run through neural networks generated in ../chembl20_mid_mwcutoff800_conformers_rms0.5_first10

### 2015-11-09 How the data was first generated:
1. **python mysql_generate_raw_data.py -a -d -c 800 ecaceres drug_matrix_chembl20_mwcutoff800.csv.gz**  
    * this file is available in our github https://github.com/keiserlab/lab-tools/tree/master/chembl  
2. Get a list of *all* unique conformers (stored in chembl20_minibatching/datasets/raw_data/from_mysql_queries/drug_mat/):  
    * **zcat drug_matrix_chembl20_mwcutoff800.csv.gz | tail -n +2 | cut -f 4 | sort | uniq > drug_mat_chembl20_mwcut800_mids.txt**  
3. Do the same for unique targets(stored in chembl20_minibatching/datasets/raw_data/from_mysql_queries/drug_mat/):  
    * **zcat drug_matrix_chembl20_mwcutoff800.csv.gz | tail -n +2 | cut -f 4 | sort | uniq > drug_mat_chembl20_mwcut800_mids.txt**  
4. Find all molecules for which conformers are not already generated by chembl20  
    * **chembl20_minibatching/datasets/raw_data/from_mysql_queries/drug_mat/drug_mat_mids_not_already_made.ipynb** 
    * .SMI written to file
        * Usually something like: zcat /tmp/all_chembl_smiles_mwcutoff800.csv.gz | awk '{print $2 " " $1}' > /srv/home/ecaceres/chembl20_minibatching/datasets/raw_data/smi_for_e3fp/all_chembl_smiles_mid_mwcutoff800.smi
        * But, in this case, since we need very few, so we print ids needed to file with the format needed SMILES\tCHEMBLID. This is also included in the .ipynb above.
5. Generate conformers:  
    * Command/set of commands  
        * /srv/home/chembl20_minibatching/datasets/e3fp_conformers/drug_matrix_chembl20_mwcutoff800_conformers_rms0.5_first10/scripts/create_conformers.sh
        * USAGE: ./create_conformers.sh smiles_file out_dir num_processors
        * HOW I USED: **./create_conformers.sh /srv/home/ecaceres/chembl20_minibatching/datasets/raw_data/smi_for_e3fp/drug_matrix/drug_mat_chembl20_mwcut800_mids_for_confs.smi /srv/home/ecaceres/chembl20_minibatching/datasets/e3fp_conformers/drug_matrix_chembl20_mwcutoff800_conformers_rms0.5_first10 25**
    * Storage location
        * /srv/home/ecaceres/chembl20_minibatching/datasets/e3fp_conformers/drug_matrix_chembl20_mwcutoff800_conformers_rms0.5_first10/level
6. Convert data to fingerprints:
    * Used /srv/home/ecaceres/chembl20_minibatching/datasets/conversion_scripts/e3fp/e3fp_format/generate_e3fp_from_sdf.sh
    * **./generate_e3fp_from_sdf.sh /srv/home/ecaceres/chembl20_minibatching/datasets/e3fp_conformers/drug_matrix_chembl20_mwcutoff800_conformers_rms0.5_first10/level/ /srv/home/ecaceres/chembl20_minibatching/datasets/ecfp_formatted_data/e3fp/drug_matrix_chembl20_mwcutoff800_conformers_rms0.5_first10/ 35**
7. Put data into hdf5 format
    * Used: /srv/home/ecaceres/chembl20_minibatching/datasets/conversion_scripts/e3fp/hdf5_format/make_drug_mat_hdf5.sh
    * see: **make_drug_mat_hdf5.sh**
